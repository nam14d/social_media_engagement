### Code to Detech Labels for the images we retrieved from the Twitter memes using Google Vision API.

```python
def detect_safe_search_uri(uri):
    """Detects unsafe features in the file located in Google Cloud Storage or
    on the Web."""
    from google.cloud import vision
    client = vision.ImageAnnotatorClient()
    image = vision.Image()
    image.source.image_uri = uri

    response = client.safe_search_detection(image=image)
    safe = response.safe_search_annotation

    # Names of likelihood from google.cloud.vision.enums
    likelihood_name = ('UNKNOWN', 'VERY_UNLIKELY', 'UNLIKELY', 'POSSIBLE',
                       'LIKELY', 'VERY_LIKELY')
    print('Safe search:')

    print('adult: {}'.format(likelihood_name[safe.adult]))
    print('medical: {}'.format(likelihood_name[safe.medical]))
    print('spoofed: {}'.format(likelihood_name[safe.spoof]))
    print('violence: {}'.format(likelihood_name[safe.violence]))
    print('racy: {}'.format(likelihood_name[safe.racy]))

    if response.error.message:
        raise Exception(
            '{}\nFor more info on error messages, check: '
            'https://cloud.google.com/apis/design/errors'.format(
                response.error.message))
        
def detect_labels(uri):
    """Detects labels in the file."""
    try:
        from google.cloud import vision
        import io
        client = vision.ImageAnnotatorClient()
        image = vision.Image()
        image.source.image_uri = uri
        response = client.label_detection(image=image)
        labels = response.label_annotations
        labelList = []
        for label in labels:
            labelList.append(str(label.description))

        if response.error.message:
            raise Exception(
                '{}\nFor more info on error messages, check: '
                'https://cloud.google.com/apis/design/errors'.format(
                    response.error.message))
    except:
        print("error")
    return labelList
```

```python
df['Labels'] = df.apply (lambda row: detect_labels(row.Image_URL), axis=1)
```

### Image Label Sentiment Scores using VaderSentiment Analysis

```python
import re
dataNew = pd.read_csv('563_df_labelled.csv')
dataNew['Labels'] = dataNew.apply (lambda row: re.findall(r'\w+', row.Labels), axis=1)
dataNew['Labels'] = dataNew.apply (lambda row: ' '.join(row.Labels), axis=1)
dataNew['Labels'] = dataNew['Labels'].astype('str')
dataNew['Labels'] = dataNew['Labels'].replace({'empty':''})
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vaderSentimentAnalyzer = SentimentIntensityAnalyzer()
dataNew['vaderSentimentScores'] = [vaderSentimentAnalyzer.polarity_scores(text)['compound'] for text in dataNew['Labels']]
dataNew['vaderSentiment'] = np.where(dataNew['vaderSentimentScores'] >= 0.05,"positive", 
                                np.where(dataNew['vaderSentimentScores'] <= -0.05,"negative",
                                         "neutral"))
```
